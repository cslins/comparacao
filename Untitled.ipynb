```python
import pandas as pd
from matplotlib import pyplot as plt
import datetime as datetime
import numpy as np
import json
import os
```

```python
pasta = "1_res_fuzzy_en0_dataset_artigo_marcos_filipe.xlsx"
excel = pd.read_excel(pasta, sheet_name='res_fuzzy_en0', header=0, engine='openpyxl')
```

```python
def save_to_npy(data, filename, kw=1):
    data['year'] = data['time'].str[:4]
    data['month'] = data['time'].str[4:6]
    data['day'] = data['time'].str[6:8]
    data['hour'] = data['time'].str[9:11]
    data['min'] = data['time'].str[11:]

    data['datetime'] = data['year'] +'-'+ data['month']+'-'+data['day']+' '+data['hour']+':'+data['min']+':00'


    data['datetime'] = pd.to_datetime(data['datetime'], dayfirst=True)
    data['date'] = data['datetime'].dt.date
    
    
    # TRANSFORMA NO FORMATO ADEQUADO O DATAFRAME PARA SER ENTRADA DA SRGAN
    lista_dates = data['date'].unique().tolist()
    
    dic = {}

    for d in lista_dates:
        dic[d] = data.loc[data['date'] == d]['P'].tolist()

    df_datesC = pd.DataFrame(dic)
    df_dates = pd.DataFrame(dic).transpose()
    
    
    # CALCULA O VALOR MAX NO DATAFRAME
    maxs = []
    for column in df_dates:
        df_dates[column] = df_dates[column]*kw
        maxs.append(df_dates[column].max())

    x_max = max(maxs)
    
    
    # TRANSFORMA TODOS OS VALORES PARA NUMEROS ENTRE 0 E 1
    for column in df_dates:
        df_dates[column] = df_dates[column]/x_max
        
    np.save("SRGAN_power_data_generation-main/data/"+filename + 'test_pv_3600', df_dates.to_numpy())

    return df_dates, x_max, lista_dates[0]

```

```python
path_of_the_directory= 'pvgis_files'
values = {}

print("Files and directories in a specified path:")
for filename in os.listdir(path_of_the_directory):
    f = os.path.join(path_of_the_directory,filename)
    if os.path.isfile(f):
        with open(f, 'r') as json_file:
            json_load = json.load(json_file)
                  
        print(f)
        df = pd.json_normalize(json_load['outputs']['hourly'])
        
        tupla = save_to_npy(df, filename, 3)
        
        values[filename] = {'max': tupla[1], 'date':tupla[2]}

        display(tupla[0])
        
        
```

```python
def load_result(filename, x_max, date):
    res = np.load(filename)
    
    
    dt = datetime.datetime(year=date.year, month=date.month, day=date.day, minute=0)
    delta = datetime.timedelta(days = 1)

    cols = []
    while dt.year != date.year+1:
        cols.append(dt)
        dt = dt + delta
        
    #cria um dataframe a partir do np.array  
    df2013 = pd.DataFrame(res, cols)
    
    # ajusta os valores no dataframe
    for column in df2013:
        df2013[column] = df2013[column]*x_max

 
    # transforma o df no formato adequado [datetime / production]
    df2013 = df2013.transpose()
    
    prod = []

    for column in df2013:
        for val in df2013[column]:
            # prod in Kwatts
            prod.append(val/1000)
 
    dt = datetime.datetime(year=date.year, month=date.month, day=date.day, minute=0)
    delta = datetime.timedelta(minutes = 5)

    cols = []
    while dt.year != date.year+1:
        cols.append(dt)
        dt = dt + delta

    dic = {'datetime': cols, 'prod': prod}
    df = pd.DataFrame(dic)
    df['day'] = df['datetime'].dt.day
    df['month'] = df['datetime'].dt.month
    df['hour'] = df['datetime'].dt.hour
    df['minute'] = df['datetime'].dt.minute
    
    
    return df
 
```

```python
path_of_the_directory= 'SRGAN_power_data_generation-main/results'

print("Files and directories in a specified path:")
for filename in os.listdir(path_of_the_directory):
    f = os.path.join(path_of_the_directory,filename)
    if os.path.isfile(f):
    
        print(f)
        
        dataframe = load_result(('SRGAN_power_data_generation-main/results/'+ filename), values[filename[:-10]]['max'],
                                values[filename[:-10]]['date'])
        
        display(dataframe)
        
        print(dataframe['prod'].max(), dataframe['prod'].mean())
        
        plt.figure(figsize=(10, 3), dpi=120)
        dataframe['prod'].plot()
        plt.show()
        
        
        name = filename.removesuffix('.jsonresult.npy')
        textfile = open(('txts/'+name + "_prod_kwatts.txt"), "w")
        
        for datas in excel['Datetime']:
            element = dataframe.loc[dataframe['datetime'] == datas].iloc[0]['prod']
            textfile.write(str(element).upper().replace('.', ',', 1) + "\n")
        textfile.close()
        '''
        rslt_df = dataframe.loc[dataframe['datetime'].isin(excel['Datetime'])]
        
        
        name = filename.removesuffix('.jsonresult.npy')
        textfile = open(('txts/'+name + "_prod_kwatts.txt"), "w")
        
        for element in rslt_df['prod']:
            textfile.write(str(element).upper().replace('.', ',', 1) + "\n")
        textfile.close()
        '''
```
