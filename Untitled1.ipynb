```python
import pandas as pd
import datetime
dfc = pd.read_excel('privatehome6.xls', sheet_name='Trend')

dfc['Hour'] = dfc['Hour'].apply(lambda x: str(x)[-8:])
dfc.loc[dfc.Hour == '00:00:00', 'Day'] = dfc.Day + datetime.timedelta(days=1)
dfc = dfc[dfc.Day.dt.year == 2012]

dfc['datetime'] = dfc['Day'].astype(str) + ' ' + dfc['Hour'].astype(str)
dfc['datetime'] = pd.to_datetime(dfc['datetime'])

dfc = pd.concat([dfc, dfc, dfc])

dfc.sort_values(by=['datetime'], inplace = True)
dfc.reset_index(drop=True, inplace=True)

d = [dfc.datetime[0]]

for i in range(366):
    for j in range(24):
        for z in range(int(60/5)):
            d.append(d[-1]+datetime.timedelta(minutes=5))
            
d.pop()

dfc.datetime = pd.Series(d)

dfc = dfc[~(dfc.Day.dt.date == datetime.date(day=29, month=2, year=2012))]

dfc.datetime = dfc.datetime.apply(lambda x: x.replace(year=2013))
```

```python
# import required module
import os
# assign directory
directory = 'result_interp'

# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        print(f)

        prod_df = pd.read_csv(f)

        prod_df.datetime = pd.to_datetime(prod_df.datetime)

        df = pd.merge(prod_df, dfc, how='inner', on='datetime')

        df = df[['datetime', 'P', 'kWh', 'kW']]

        df.P = df.P.apply(lambda x: x/1000)

        df['El_KW'] = df.P - df.kW

        df['Hora'] = df.datetime.apply(lambda x: x.hour+x.minute/60.0)

        df.to_excel('dfs_para_scripts/'+filename+'.xlsx')
        
        display(df)
```
