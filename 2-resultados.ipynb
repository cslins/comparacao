```python
import pandas as pd
import datetime

pasta = "original.xlsx"
excel = pd.read_excel(pasta, sheet_name='res_fuzzy_en0', header=0, engine='openpyxl')
```

```python
excel.rename(columns={'Datetime': 'datetime'}, inplace=True)
```

```python
# import required module
import os
# assign directory
directory = 'result_interp'

# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        print(f)

        prod_df = pd.read_csv(f)

        prod_df.datetime = pd.to_datetime(prod_df.datetime)

        df = pd.merge(prod_df, excel, how='right', on='datetime')

        df = df[['datetime', 'P', 'Ec_KWh', 'Ec_KW', 'Hora']]

        df.P = df.P.apply(lambda x: x/1000)

        df['El_KW'] = df.P - df.Ec_KW

        # df['Hora'] = df.datetime.apply(lambda x: x.hour+x.minute/60.0)

        df.to_excel('dfs_para_scripts/'+filename+'.xlsx')
        
        display(df)
```
